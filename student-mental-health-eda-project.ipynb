```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
import numpy as np
```

```python
df = pd.read_csv("/kaggle/input/student-mental-health-survey/MentalHealthSurvey.csv")
```

```python
df.info()
```

```python
# Checking for missing data
df.isnull().sum()
```

```python
# Gender Distubution Pie chart of suvery responders
gender_counts = df['gender'].value_counts()

plt.figure(figsize=(8, 5))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Gender Distribution')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular
plt.show()
```

```python
# Descriptive Statistics 
pd.set_option('display.float_format', lambda x: '%.2f' % x )
df.describe()
```

```python
# Finding the Strongest Correlations between numerical collums

# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=[np.number])

# Calculate the correlation matrix using only numeric columns
correlation_matrix = numeric_df.corr()

# Display the correlation matrix
print(correlation_matrix)

# Get the absolute values of the correlation matrix and drop self-correlations
strongest_correlation = correlation_matrix.abs().unstack().sort_values(ascending=False)

# Exclude self-correlations
strongest_correlation = strongest_correlation[strongest_correlation < 1]

# Get the strongest correlation pair
top_correlation = strongest_correlation.idxmax()
print(f'The strongest correlation is between {top_correlation[0]} and {top_correlation[1]}: {strongest_correlation.max():.2f}')
```

```python
# Top 3 highsest correlations

# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=[np.number])

# Calculate the correlation matrix using only numeric columns
correlation_matrix = numeric_df.corr()

# Get the absolute values of the correlation matrix and drop self-correlations
strongest_correlation = correlation_matrix.abs().unstack().sort_values(ascending=False)

# Exclude self-correlations
strongest_correlation = strongest_correlation[strongest_correlation < 1]

# Get the top 5 correlation pairs
top_5_correlations = strongest_correlation.head(5)

# Display the top 5 strongest correlations, but only the 1st, 3rd, and 5th
print("Top 3 selected correlations (1st, 3rd, and 5th):")
for i, ((col1, col2), corr_value) in enumerate(top_5_correlations.items(), start=1):
    if i in [1, 3, 5]:
        print(f"{col1} and {col2}: {corr_value:.2f}")
```
